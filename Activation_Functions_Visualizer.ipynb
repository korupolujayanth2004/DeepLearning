```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

def binary_step(x):
    return np.where(x >= 0, 1, 0)

def linear(x):
    return x

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.plot(x, binary_step(x), color='purple')
plt.title('Binary Step')
plt.grid(True)

plt.subplot(3, 2, 2)
plt.plot(x, linear(x), color='blue')
plt.title('Linear')
plt.grid(True)

plt.subplot(3, 2, 3)
plt.plot(x, relu(x), color='green')
plt.title('ReLU')
plt.grid(True)

plt.subplot(3, 2, 4)
plt.plot(x, sigmoid(x), color='red')
plt.title('Sigmoid')
plt.grid(True)

plt.subplot(3, 2, 5)
plt.plot(x, tanh(x), color='orange')
plt.title('Tanh')
plt.grid(True)

plt.subplot(3, 2, 6)
x_softmax = np.vstack([x, x + 2])
softmax_vals = softmax(x_softmax)
plt.plot(x, softmax_vals[0], label='Class 1')
plt.plot(x, softmax_vals[1], label='Class 2')
plt.title('Softmax')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
